```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings to display all columns
pd.set_option("display.max_columns", None)

# Load the datasets
train_df = pd.read_csv('../data/raw/coding_challenge_train.csv')
test_df = pd.read_csv('../data/raw/coding_challenge_test_without_labels.csv')

# Display the first few rows of the training dataset
train_df.head()

# Summary statistics for the training dataset
train_df.describe()

# Check for missing values in the training dataset
train_df.isnull().sum()

# Visualizations

# Histograms for all numerical features in the training dataset
train_df.hist(bins=30, figsize=(20, 15), layout=(7, 5))
plt.show()

# Boxplots for all numerical features to identify outliers
num_features = train_df.select_dtypes(include=[np.number]).columns.tolist()
train_df[num_features].plot(kind='box', subplots=True, layout=(7, 5), figsize=(20, 15))
plt.show()

# Correlation matrix heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(train_df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.show()

# Scatter plots for the main features correlated with the target
# Replace 'target_regression' and 'target_classification' with the actual names of your target columns
sns.scatterplot(data=train_df, x='feature_1', y='target_regression')
plt.show()

sns.scatterplot(data=train_df, x='feature_1', y='target_classification')
plt.show()

# Count plot for the classification target
sns.countplot(x='target_classification', data=train_df)
plt.show()

# Analysis of categorical features
# Assume 'categorical_feature' is the name of a categorical column in your dataset
sns.countplot(x='categorical_feature', data=train_df)
plt.xticks(rotation=45)
plt.show()

# Pairplot for selected features
# Select features relevant to your regression and classification target and replace them in the list below
sns.pairplot(train_df[['feature_1', 'feature_2', 'target_regression', 'target_classification']])
plt.show()

# Any additional analysis that might be required
# ...
```
